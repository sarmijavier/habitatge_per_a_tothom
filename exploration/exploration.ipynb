```python
import pandas as pd
```

```python
df = pd.read_csv('estat_tipsho10.tsv', sep='\t')
```

```python
df
```

```python
import matplotlib.pyplot as plt

# 1. Isolate the first row and the year columns
# We skip the first column ('freq,unit,geo...') and take all year columns
first_country = df.iloc[10, 11:] 

# 2. Data Cleaning: Remove flags like 'e' or 'b' and convert to float
# This ensures "80.00 e" becomes 80.00
clean_data = first_country.astype(str).str.replace(r'[a-zA-Z]', '', regex=True).astype(float)

# 3. Plotting
plt.figure(figsize=(10, 5))
plt.plot(clean_data.index, clean_data.values, marker='o', linestyle='-', color='royalblue')

plt.title('Trend for Austria (AT)')
plt.xlabel('Year')
plt.ylabel('Value')
plt.xticks(rotation=45) # Tilt years for better readability
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
